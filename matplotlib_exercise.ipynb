```python
# 使うライブラリ
# !pip install japanize-matplotlib

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()
```

## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
data = pd.read_csv('./hist_data.csv')

a_data = data['aデータ']
b_data = data['bデータ']

plt.hist(a_data, bins=30, alpha=0.5, label='aデータ', color='orange')
plt.hist(b_data, bins=30, alpha=0.5, label='bデータ', color='green')

plt.xlabel('温度')
plt.ylabel('度数')
plt.title('aデータとbデータの温度に関する度数分布')
plt.legend(loc='upper right')
plt.grid(False)

plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
data = pd.read_csv('./scatter_data.csv')

class0 = data[data['class'] == 0]
class1 = data[data['class'] == 1]
class2 = data[data['class'] == 2]

plt.scatter(class0['length (cm)'], class0['width (cm)'], label='Class 0', color='red')
plt.scatter(class1['length (cm)'], class1['width (cm)'], label='Class 1', color='blue')
plt.scatter(class2['length (cm)'], class2['width (cm)'], label='Class 2', color='green')

plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.title('クラス0~2のlengthとwidthに関する散布図')
plt.legend(loc='upper right')
plt.grid(True)

plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
data = pd.read_csv('./line_data.csv')

data.set_index('期間', inplace=True)

a_data = data['件数']

plt.plot(a_data, marker='o')

plt.xlabel('期間')
plt.ylabel('件数')
plt.title('期間ごとの件数の推移')
plt.grid(True)


plt.xticks(rotation=30)

plt.show()




data = pd.read_csv('./line_data.csv')

period = data['期間']
b_data = data['費用']

plt.bar(period, b_data)

plt.xlabel('期間')
plt.ylabel('費用(円)')
plt.title('期間ごとの費用の推移')
plt.grid(True)

plt.xticks(rotation=30)

plt.show()
```
